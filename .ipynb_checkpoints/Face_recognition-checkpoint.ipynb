2) Preprocessing , scaling and resizing the image 


```python
def preprocess(file_path) :
    byte_img = tf.io.read_file(file_path) 
    img = tf.io.decode_jpeg(byte_img) 
    # resizing the image 
    img = tf.image.resize(img, (100,100)) 
    # normalizing the image 
    return img/255.0
```

3) Creating a labelled dataset

   
    (anchor , positive) => 1.1.1.1.1.1.1 (upto 300) <br>
    (anchor, negative) => 0.0.0.0.0.0.0 (upto 300 )

   we have created a set of images and labelled the sets as 0 and 1 , which is the    requirement for data in siamese network 

```python
'''
-> len(anchor) -> This computes the number of elements in the anchor dataset
-> tf.ones(len(anchor) -> tf.ones(len(anchor)): Suppose len(anchor) is 100. This creates a tensor of shape (100,) filled with ones: [1.0, 1.0, ..., 1.0] (100 times)
-> tf.data.Dataset.from_tensor_slices(tf.ones(len(anchor))) 
    *) If you pass a tensor [1.0, 1.0, ..., 1.0] (of shape (100,)), this function will create a dataset with 100 elements, each being 1.0.

-> tf.data.Dataset.zip((anchor, positive, tf.data.Dataset.from_tensor_slice(tf.ones(len(anchor))))): 
    *) The zip function combines these three datasets (anchor, positive, and the dataset of ones) into a single dataset. 
    *) Each element in the new zipped dataset is a tuple where the first element is from the anchor dataset, the second element is from the positive dataset, 
       and the third element is from the dataset of ones

       e.g -> (b'data\\anchor\\65c50ccf-18e8-11ef-ad6d-202b2082a7be.jpg', b'data\\positive\\82e7aa01-18e8-11ef-a623-202b2082a7be.jpg', 1.0)
'''

positives = tf.data.Dataset.zip((anchor, positive , tf.data.Dataset.from_tensor_slices(tf.ones(len(anchor)))))
negatives = tf.data.Dataset.zip((anchor, negative , tf.data.Dataset.from_tensor_slices(tf.zeros(len(anchor)))))
data = positives.concatenate(negatives)
```

# Building the model 

**Below we have created a model which would generate the embeddings for our dezired images which will be checked by us**

```python
def make_embedding() :
    inp = Input(shape = (100, 100, 3) , name = 'input_image') 
    c1 = Conv2D(64 , (10,10), activation = 'relu')(inp) 
    m1 = MaxPooling2D(64 , (2,2) , padding = 'same')(c1) 
    c2 = Conv2D(128, (7,7) , activation = 'relu')(m1) 
    m2 = MaxPooling2D(64 , (2,2), padding='same')(c2)
    c3 = Conv2D(128 , (4,4) , activation = 'relu')(m2)
    m3 = MaxPooling2D(64 , (2,2), padding = 'same')(c3)
    c4 = Conv2D(256, (4,4) , activation = 'relu')(m3)
    f1 = Flatten()(c4)
    d1 = Dense(4096 , activation = 'sigmoid')(f1)
    # through below return statement I am returning a model with all the above layer structure by passing it to a model class which I have imported 
    # earlier
    return Model(inputs = [inp] , outputs = [d1] , name = 'embedding')
```

```python
embedding = make_embedding()
```

```python
embedding.summary()
```

**Now I want to compare the embeddings generated by the above model for anchor image as well as the negative/positive image to know the differance between them which would help us to know that how similar or dissimar our images are** <br>
**In siamese research paper there is a L1 distance layer , so now I am going to create a L1 siamese distance layer** 

```python
class L1Dist(Layer) :
    def __init__(self, **kwargs) :
        super().__init__()
    def call(self , input_embedding , validation_embedding) :
        return tf.math.abs(input_embedding - validation_embedding)
```

```python
l1 = L1Dist()
```

```python
l1
```

**Now we are going to combine the l1 layer and the embedding part together**

```python
def make_siamese_model() :
    input_image = Input(name = 'input_img' , shape = (100,100,3))
    validation_image = Input(name = 'validation_img' , shape = (100,100,3))
    siamese_layer = L1Dist()
    siamese_layer._name = 'distance'
    distances = siamese_layer(embedding(input_image) , embedding(validation_image))
    # classifier to know whether this image belongs to the same person or not
    classifier = Dense(1 , activation = 'sigmoid')(distances)
    return Model(inputs = [input_image, validation_image], outputs = classifier , name = 'SiameseNetwork')
```

```python
siamese_model = make_siamese_model()
```

```python
siamese_model.summary()
```

```python
# Building a train step function -> These steps are same for any model that we build
@tf.function # this function will compile the below function and will help us to train our 
# model in an efficient manner 
def train_step(batch) :
    with tf.GradientTape() as tape :
        # Get anchor and positive /negative images
        X = batch[:2] 
        # Get label 
        y = batch[2]
        # forward pass 
        yhat = siamese_model(X , training=True) 
        # loss calculation 
        loss = binary_cross_loss(y , yhat)
    print(loss)    

    grad = tape.gradient(loss , siamese_model.trainable_variables)
    # calculating updated weights and apply to siamese model
    opt.apply_gradients(zip(grad, siamese_model.trainable_variables))
    return loss 
```

```python
# building training loop 
def train(data, EPOCHS) :
    for epoch in range(1, EPOCHS+1) :
        print('\n Epoch {}/{}'.format(epoch, EPOCHS))
        progbar = tf.keras.utils.Progbar(len(train_data))
        # loop through batch
        for idx , batch in enumerate(train_data) :
            train_step(batch) 
            progbar.update(idx+1)

        if epoch % 10 ==0 :
            checkpoint.save(file_prefix=checkpoint_prefix)


```

# Real time test

Now in general our basic approach is that we will take our input image and pass that image to our model and similarly take a positive image and pass that image too into our model and then compare their embeddings . <br>
But here instead of passing only one image for verification , we will pass 50 verification images and will check the similarity of imput image with each if those 50 images . Then we will make a decision threshold i.e suppose the decision threshold is 30/50 = 0.6 , it means that if the ratio of same images and total images is less than 0.6 , we will say that the image is not verified <br>

We created a folder named application data and in this folder I have created two folders:
   1.  input folder
   2.  verification folder<br>
verification folder contains in total 50 images. We will be going to verify our imput images with each of the 50 images and then will use decision threshold to predict our result

## verification function

```python
os.path.join('application data' , 'verification_images') # it will create a path as shown below 
```

```python
# now to access path of all the images in this directory we will use os.listdir() function 
os.listdir(os.path.join('application data' , 'verification_images'))
# it is listing all the images which is present in our verification_images folder 
```

```python
def verify(frame , model , detection_threshold , verification_threshold) :
    results = []
    for image in os.listdir(os.path.join('application data', 'verification_images')) :
        input_image = preprocess(os.path.join('application data', 'input_image' , 'input_image.jpg'))
        validation_img = preprocess(os.path.join('application data', 'verification_images', image))
        result = model.predict(list(np.expand_dims([input_image , validation_img], axis = 1)))
        results.append(result)
    detection = np.sum(np.array(results)>detection_threshold) 
    verification = detection / len(os.listdir(os.path.join('application data', 'verification_images')))
    verified = verification > verification_threshold 
    return results , verified
```

```python
import cv2
import os
#import uuid

# Define paths for the anchor and positive images
# ANC_PATH = 'path_to_anchor_images'
# POS_PATH = 'path_to_positive_images'
# Open the camera at index 0
cap = cv2.VideoCapture(0)

# Check if camera is opened successfully
if not cap.isOpened():
    print("Error: Could not open camera.")
    exit()

while cap.isOpened():
    # frame contains the image which is collected from the webcam 
    ret, frame = cap.read()
    # making frame size having dimensions equal to the negative images dimension k
    frame = frame[120:120+250, 200:200+250, :] 

    # Check for key events
    key = cv2.waitKey(1) & 0xFF
    
    if key == ord('t'):
        #imgname = os.path.join('application data', 'input_image', 'input_image.jpg'.format(uuid.uuid1()))
        #temp = imgname
        cv2.imwrite(os.path.join('application data', 'input_image', 'input_image.jpg'), frame)
        results , verified = verify(frame , model , 0.5, 0.5)
        print(verified)
        
    if key == ord('q'):
        break
    
    cv2.imshow('Image Collection', frame)

cap.release()
cv2.destroyAllWindows()
```
